```python
import input_pipeline
import MyKeras
import matplotlib.pyplot as plt
import numpy as np
from keras.utils import to_categorical
```

```python
image_shape = (288, 512,3)
data_path = '/home/tsou/Desktop/ICME2023/iVS-ODSEG-Dataset'
train_gen , valid_gain= input_pipeline.get_generators(batch_size=1, dataset_path=data_path,image_shape=image_shape,real_word=False)
```

```python
model, epoch = MyKeras.load_latest_model('models/main')
```

```python
i = 500
for x, y in train_gen:
    i+=1
    y_hat = model.predict(x)
    y_hat = np.argmax(y_hat, axis=-1)
    y_hat = np.array(to_categorical(y_hat, 3)[0] * 255.0).astype(np.uint8)
    y_hat[:,:,0] = 0

    # plt.imshow(np.array(x[0]*255.0 + y_hat/2.0).astype(np.uint8))
    plt.imsave('predict_result/images/'+str(i)+'.png',np.array(x[0]*255.0 + y_hat/2.0).astype(np.uint8))
    # plt.show()
```
